```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage.color import rgb2gray
from PIL import Image
import pandas as pd
import numpy as np
import cv2
import colorsys
```

```python
original=Image.open("yoda.jpeg")
grayscale=Image.open("yoda.jpeg").convert('LA')
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(original)
ax[0].set_title("Original")
ax[1].imshow(grayscale, cmap=plt.cm.gray)
ax[1].set_title("Grayscale")

fig.tight_layout()
plt.show()
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.3, 0.587, 0.144])
```

```python
img = mpimg.imread('yoda.jpeg')

gray = rgb2gray(img)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(img)
ax[0].set_title("Original")
ax[1].imshow(gray, cmap=plt.cm.gray)
ax[1].set_title("Grayscale")

fig.tight_layout()
plt.show()
```

```python
yoda=cv2.imread("yoda.jpeg")
yoda_array=np.asarray(yoda)
yoda2=cv2.imread("yoda2.jpg")
yoda2_array=np.asarray(yoda2)
print(yoda_array)
print(yoda2_array)
```

```python
weightedSum = np.add(yoda_array,yoda2_array)
plt.imshow(weightedSum)
plt.show()
```

```python
sub = np.subtract(yoda_array,yoda2_array)
plt.imshow(sub)
plt.show()
```

```python
mul=np.multiply(yoda_array,yoda2_array)
plt.imshow(mul)
plt.show()
```

```python
hsv_img = cv2.cvtColor(yoda, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_img)
plt.show()
```

```python
#nearest neighbour
near_img = cv2.resize(yoda,None, fx = 10, fy = 10, interpolation = cv2.INTER_NEAREST)
plt.imshow(near_img)
plt.show()
```

```python
#Bilinear
bilinear_img = cv2.resize(yoda,None, fx = 10, fy = 10, interpolation = cv2.INTER_LINEAR)
plt.imshow(bilinear_img)
plt.show()
```

```python
bicubic_img = cv2.resize(yoda,None, fx = 10, fy = 10, interpolation = cv2.INTER_CUBIC)
plt.imshow(bicubic_img)
plt.show()
```
