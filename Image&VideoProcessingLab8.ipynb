```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.util import random_noise
```

```python
img = cv2.imread('dataset/j.png',0)
kernel = np.ones((5,5),np.uint8)
plt.imshow(img)
```

```python
#erosion
erosion = cv2.erode(img,kernel,iterations = 1)
plt.imshow(erosion)
```

```python
#dilation
dilation = cv2.dilate(img,kernel,iterations = 1)
plt.imshow(dilation)
```

```python
#Adding salt and pepper noise
noise_img = random_noise(img, mode='s&p',amount=0.03)
plt.imshow(noise_img)
```

```python
#opening
opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernel)
plt.imshow(opening)
```

```python
#closing
closing = cv2.morphologyEx(noise_img, cv2.MORPH_CLOSE, kernel)
plt.imshow(closing)
```

```python
noisy_img=cv2.imread('dataset/mandrill_noisy.png',0)
plt.imshow(noisy_img)
```

```python
#opening
opening = cv2.morphologyEx(noisy_img, cv2.MORPH_OPEN, kernel)
plt.imshow(opening)
```

```python
#closing
closing = cv2.morphologyEx(noisy_img, cv2.MORPH_CLOSE, kernel)
plt.imshow(closing)
```

```python
#opening and closing both
opening = cv2.morphologyEx(noisy_img, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel)
plt.imshow(closing)
```

```python
#closing and opening
closing = cv2.morphologyEx(noisy_img, cv2.MORPH_CLOSE, kernel)
opening = cv2.morphologyEx(closing, cv2.MORPH_OPEN, kernel)
plt.imshow(opening)
```
