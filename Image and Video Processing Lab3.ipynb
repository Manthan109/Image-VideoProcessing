```python
import pandas as pd
import numpy as np
from PIL import Image
from numpy import asarray
from matplotlib import pyplot
import glob
import cv2
```

```python
#converting image to numpy array
image=Image.open("dataset/yoda.jpeg")
data=asarray(image)
data
```

```python
#converting numpy array back to image
array_conversion=Image.fromarray(data)
print(type(array_conversion))
pyplot.imshow(array_conversion)
```

```python
#saving image from jpeg to png
format_conversion=Image.fromarray(data).save('yoda_format_modified.png')
```

```python
#cropping image using python library
image_crop=image.crop((200,100, 400, 400))
pyplot.imshow(image_crop)
```

```python
#cropping image using numpy array
cropped=data[128:560,300:800,:]
pyplot.imshow(cropped)
data.shape
```

```python
#negative of image
negative_image=255-data
pyplot.imshow(negative_image)
```

```python
#reading and plotting multiple images
folder=glob.glob("dataset/cats")
image_list=[]
for i in folder:
    for f in glob.glob(i+'/*.jpg'):
        image_list.append(f)

read_images = []        
for image in image_list:
    read_images.append(cv2.imread(image))
```

```python
f=pyplot.figure()
for i in range(len(read_images)):
    f.add_subplot(1,len(read_images),i+1)
    pyplot.imshow(read_images[i])
    
```

```python
#creating circular mask on the input image
im=cv2.imread("dataset/yoda2.jpg")
circle=cv2.imread("dataset/circle.png")
circle=cv2.resize(circle,im.shape[1::-1])
```

```python
dst = cv2.bitwise_and(im,circle)
```

```python
pyplot.imshow(dst)
```
